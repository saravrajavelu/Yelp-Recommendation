```python
# Library Imports 

import numpy as np
import pandas as pd
import pickle

% matplotlib inline
```

```python
#business = pd.read_json('./Data/business.json', lines =  True)
business = pd.read_pickle('business.pkl')

print(business.shape)
business.head()
```

```python
review = pd.DataFrame()

df = pd.read_json('./Data/review.json',lines = True, chunksize= 1e5)

try:
   for df_chunk in df:
       review = pd.concat([review, df_chunk])
except ValueError:
       print ('\nSome messages in the file cannot be parsed')


print(review.shape)

review.head()
```

```python
# Dropping non restaurants and non-food businesses
bus_cat_filter = business.categories.apply(lambda x: False if 'Restaurants' in x or 'Food' in x else True)
business.drop(business[bus_cat_filter].index, inplace = True)
```
